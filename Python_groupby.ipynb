```python
us_jobs = df[df['job_country']=='United States']

us_jobs
```

```python
us_jobs = us_jobs[us_jobs['salary_year_avg'].notna()]
```

```python
us_jobs.groupby(['job_title_short'])[['salary_year_avg']].agg(['mean','min','max','count']).sort_values(by=('salary_year_avg', 'mean'), ascending=False)
```
